```python
import pandas as pd
import yfinance as yf
```

```python
end_date = pd.Timestamp.today().normalize()

symbols = [
    # Precious metals
    "GC=F",  # Gold
    "SI=F",  # Silver
    "PL=F",  # Platinum
    "PA=F",  # Palladium

    # Energy
    "CL=F",  # WTI Crude
    "BZ=F",  # Brent
    "NG=F",  # Natural Gas
    "RB=F",  # Gasoline
    "HO=F",  # Heating Oil

    # Industrial metals (only what usually works)
    "HG=F",  # Copper

    # Grains
    "ZC=F",  # Corn
    "ZW=F",  # Wheat
    "ZS=F",  # Soybeans

    # Softs
    "KC=F",  # Coffee
    "CC=F",  # Cocoa
    "SB=F",  # Sugar
    "CT=F",  # Cotton

    # Livestock
    "LE=F",  # Live Cattle
    "HE=F",  # Lean Hogs
]

df = yf.download(
    tickers=symbols,
    start="1990-01-01",
    end=end_date,
    interval="1d",
    auto_adjust=False,
    group_by="ticker",
    threads=True
)
```

```python
df.tail()
```

```python
df.to_parquet("commodities_data.parquet")
```

```python
df_close = df.xs("Close", level=1, axis=1)

df_close.to_parquet("commodities_close.parquet", engine="pyarrow")
```
